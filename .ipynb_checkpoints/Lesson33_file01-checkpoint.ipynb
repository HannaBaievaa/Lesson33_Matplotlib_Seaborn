#імпортуємо необхідні бібліотеки та додаємо наш excel файл

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
avto = pd.read_excel("AVTO.xlsx")
```

#### Візуалізуємо залежність споживання палива від вартості за допомогою лінійної діаграми

```python
sns.set_style('white')
sns.set_palette('deep')
sns.lineplot(data=avto,x='Споживання палива, л на 100 км',y='Вартість, грн')
```

#### Візуалізуємо залежність споживання палива від макс швидкості за допомогою точкового графіку

```python
sns.set_style('white')
sns.set_palette('deep')
sns.scatterplot(data=avto,x='Споживання палива, л на 100 км',y='Макс швидкість, км/год')
```

#### Візуалізуємо залежність вантажопідйомності від макс швидкості за допомогою точкового графіку

```python
sns.set_style('darkgrid')
sns.set_palette('deep')
sns.scatterplot(data=avto,x='Вантажопідйомність, кг',y='Макс швидкість, км/год')
```

#### Побудуємо гістограму для розділення автомобілів на 5 груп за швидкістю за допомогою бібліотеки seaborn

```python
sns.set_style('darkgrid')
sns.set_palette('deep')
sns.displot(data=avto,x='Марка авто',bins=5, kde=True)
```

#### Побудуємо гістограму для розділення автомобілів на 5 груп за швидкістю за допомогою бібліотеки matplotlib

```python
speed =[150,150,142,142,155,150,150,157,150,183,173,160,188,157,165,150,150,170,155,155]
speed_sorted = sorted(speed)
```

```python
import matplotlib.pyplot as plt
plt.hist(speed_sorted, 5, color = "green")
```

#### Побудуємо коробковий графік для відображення швидкостей різних марок авто

```python
sns.boxplot(data=avto, x='Макс швидкість, км/год')
```

#### Побудуємо парний графік для відображення залежностей величин датасет

```python
sns.pairplot(data=avto)
```
